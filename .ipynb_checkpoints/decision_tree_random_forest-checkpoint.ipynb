```python
import pandas as pd

df = pd.read_csv("titanic-passengers.csv", sep=";")
df.head()
```

```python
#CLEANING
df.drop(["Cabin", "PassengerId", "Ticket", "Name"], axis=1, inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

#TRANSFORMATION
cleanup_nums = {
    "Survived": {"No": 0, "Yes": 1},
    "Sex": {"male": 0, "female": 1},
    "Embarked": {"C": 0, "Q": 1, "S": 2}
}
df.replace(cleanup_nums, inplace=True)
```

# 1. Applying decision tree

```python
from sklearn.model_selection import train_test_split   
from sklearn import tree
from sklearn.metrics import accuracy_score

#features extraction
X = df[['Sex','Pclass','Age']]
Y = df["Survived"]

#splitting data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=10)

#applying tree algorithm
my_tree = tree.DecisionTreeClassifier(random_state=0, criterion="gini", splitter="random", max_depth=6)
my_tree.fit(X_train, Y_train)
Y_pred = my_tree.predict(X_test)
print(f"score: {accuracy_score(Y_test, Y_pred)}")
```

# 2. Plotting decision tree

```python
import graphviz
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files/Graphviz/bin/'

dot_data = tree.export_graphviz(my_tree, feature_names=['Sex','Pclass','Age'], class_names=['Dead', 'Survived'], out_file=None, filled=True)
graph = graphviz.Source(dot_data)
graph
```

If we take a 26 years old man who travel in Pclass 1, we can predict with the help of the decision tree that he will survive. 

# 3. Change the decision tree parameters

```python
my_tree = tree.DecisionTreeClassifier(random_state=0, criterion="entropy", splitter="best", max_depth=6)
my_tree.fit(X_train, Y_train)
```

# 4. Calculate the new accuracy and compare it with the previous results

```python
Y_pred = my_tree.predict(X_test)
print(f"score: {accuracy_score(Y_test, Y_pred)}")
```

We notice a decrease of the accuracy when we change criterion from "gini" to "entropy" and splitter from "random" to "best".

# 5. Random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_tree=RandomForestClassifier(n_estimators=10)
rf_tree.fit(X_train, Y_train)
Y_pred=rf_tree.predict(X_test)
print(f"Accuracy: {accuracy_score(Y_test, Y_pred)}")
```

```python
rf_tree=RandomForestClassifier(n_estimators=20)
rf_tree.fit(X_train, Y_train)
```

# 6. Calculate the new accuracy and compare it with the previous result

```python
Y_pred=rf_tree.predict(X_test)
print(f"Accuracy: {accuracy_score(Y_test, Y_pred)}")
```

We notice that changing the number of estimators from 10 to 20, weakly affects the accuracy.
